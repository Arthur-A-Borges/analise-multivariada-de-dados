IMPORTAÇÃO DE DADOS:

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer

dados_cancer = load_breast_cancer()
```

CRIAÇÃO DE TABELAS:

```python
df = pd.DataFrame(dados_cancer.data, columns=dados_cancer.feature_names)


df['diagnosticos'] = dados_cancer.target
```

VISUALIZAÇÃO DE TABELA:

```python
df.info()
```

DADOS EM PORCENTAGEM(CASOS DE CANCER BENIGNO E MALIGNO):

```python
porcentagem = df['diagnosticos'].value_counts(normalize=True) * 100
print(porcentagem)
```

TENATIVA DE DROP EM UMA COLUNA:

```python
df = df.drop('id', 'nome', axis = 1)
```
